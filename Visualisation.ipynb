```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("survey_data.csv", header=0)


# Visualization 1: Age with Highest Income
age_groups = data.groupby("Age")["Total Income"].mean()
plt.bar(age_groups.index, age_groups.values)
plt.xlabel("Age Group")
plt.ylabel("Average Income")
plt.title("Average Income by Age Group")
plt.show()

# Visualization 2: Gender Distribution of Spending
gender_spending = data.groupby(["Gender", "Utilities", "Entertainment", "School Fees", "Shopping", "Healthcare"]).sum()
gender_spending = gender_spending.unstack()
gender_spending.plot(kind="bar", stacked=True)
plt.xlabel("Expense Category")
plt.ylabel("Total Spending")
plt.title("Spending Distribution by Gender")
plt.legend(title="Gender")
plt.show()

# Export charts for presentation
# plt.savefig("age_income.png")
# plt.savefig("gender_spending.png")
```
